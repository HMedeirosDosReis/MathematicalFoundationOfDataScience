```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_lfw_people
```

```python
#Each row corresponds to a ravelled face image of original size 62 x 47 pixels. 
faces = fetch_lfw_people(min_faces_per_person=60)
print(type(faces))
```

```python
y = faces.target
X = faces.data

print(y.shape)
print(X.shape)
X[0, 0:4]
```

```python
names = faces.target_names
print(names)
```

```python
images = faces.images
print(images.shape)
```

```python
#show random images from array "images".  Run this cell several times to see them!
i = np.random.choice(range(images.shape[0]))
plt.close()
plt.rcParams["figure.figsize"] = [4,4]
plt.imshow(images[i], cmap='bone')
plt.axis('off')
plt.show()
```

```python
_, name_dist = np.unique(y, return_counts=True)

plt.close()
plt.bar(range(len(names)), name_dist)
plt.xticks(range(len(names)), names, rotation='vertical')
plt.show()
```

## Train-Test Splits and Rescaling

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = 0.20, random_state = 1)

print(y_train.size)
print(y_val.size)
```

Here we scale X.  Note that the values of x range from 0 to 255 (the shades of gray).

```python
Xs = X / 255
Xs_train = X_train / 255
Xs_val = X_val / 255
```

## Import Algorithms

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
```

## Logistic Regression

```python
%%time
lr_mod = LogisticRegression(solver='lbfgs', multi_class='auto', max_iter=2000)
lr_mod.fit(Xs_train, y_train)
print(lr_mod.score(Xs_train, y_train))
print(lr_mod.score(Xs_val, y_val))
```

```python
i = np.random.choice(range(X_val.shape[0]))
pred = lr_mod.predict([Xs_val[i,:]])
plt.imshow(X_val[i].reshape(62,47), cmap='bone')
plt.axis('off')
plt.text(50, 5, s = names[y_val[i]], fontsize=16, color='b')
plt.text(50, 15, s = names[pred[0]], fontsize=16, color='r')
plt.show()
```

```python
pred_val = lr_mod.predict(Xs_val)
```

```python
print( classification_report(y_val, pred_val, target_names=names))
```

```python
print(confusion_matrix(y_val, pred_val))
```

## KNN

```python
%%time
KNN_mod = KNeighborsClassifier(n_neighbors=12)
KNN_mod.fit(Xs_train, y_train)
print(KNN_mod.score(Xs_train, y_train))
print(KNN_mod.score(Xs_val, y_val))
```

```python
pred_val = KNN_mod.predict(Xs_val)
print( classification_report(y_val, pred_val, target_names=names))
print(confusion_matrix(y_val, pred_val))
```

## SVM

```python
%%time
svm_mod = SVC(kernel='rbf', gamma=0.002, C=10)
svm_mod.fit(Xs_train, y_train)
print(svm_mod.score(Xs_train, y_train))
print(svm_mod.score(Xs_val, y_val))
```

```python
pred_val = svm_mod.predict(Xs_val)
print( classification_report(y_val, pred_val, target_names=names))
print(confusion_matrix(y_val, pred_val))
```

# PCA

Now we will use the Principal Component Analysis functions to see if we can keep the accuracy of the predictions while cutting down the TRAINING TIME.   We begin by selecting 150 components out of 2914 features in the data set.

```python
from sklearn.decomposition import PCA
```

```python
%%time
pca = PCA(n_components = 150, whiten=True)
Z = pca.fit_transform(X)
print(np.shape(Z))
```

```python
print(np.cumsum(pca.explained_variance_ratio_))
```

```python
pc = pca.components_
print(np.shape(pc))
np.shape(pc[1])
np.shape(Z)
```

```python
#these are the first 25 faces in the transformed space
#thus we show the first 25 composite faces saved in np.array called "pc"
plt.close()
plt.rcParams["figure.figsize"] = [12,12]
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.imshow(pc[i].reshape(62,47), cmap='bone')
    plt.axis('off')
plt.show()
```

```python
#here we transform back to the original X 
#with one vector Z[0,i] and one vector pc[i] at a time.
#we do it for the first 25 vectors pc[i]
cp = np.zeros(len(pc[0])).reshape(62,47)

plt.close()
plt.rcParams["figure.figsize"] = [12,12]
for i in range(0, 25):
    #here is the reconstruction: from Z back to X
    #one step at the time
    cp += Z[0,i] * pc[i].reshape(62,47)
    
    plt.subplot(5,5,i+1)
    plt.imshow(cp, cmap='bone')
    plt.axis('off')
plt.show()
```

```python
#here we transform back to the original X
#like in the cell above, but we do it for all 150 principal components pc[i]
cp = np.zeros(len(pc[0])).reshape(62,47)

plt.close()
plt.rcParams["figure.figsize"] = [12,12]
for k in {0,25,50,75, 100, 125}:
   temp=k
   for i in range(0+temp, 25+temp):
    
     cp += Z[0,i] * pc[i].reshape(62,47)
    
     plt.subplot(5,5,i+1 - temp)
     plt.imshow(cp, cmap='bone')
     plt.axis('off')
   plt.show()
```

```python
# here we compare the reconstructed face from 150 of  Z[0,i] and pc[i]
#with the original face from images[0]
cp = Z[0,:].reshape(150,1) * pc[:,:]
cp = np.sum(cp, axis=0)

plt.rcParams["figure.figsize"] = [2,2]
plt.imshow(cp.reshape(62,47), cmap='bone')
plt.axis('off')
plt.show()

plt.rcParams["figure.figsize"] = [2,2]
plt.imshow(images[0], cmap='bone')
plt.axis('off')
plt.show()
```

After seeing how only 150 principal components saved in "pc" can be used to recreate recognizably a picture created with 2948 original components (features), we now turn to using the PCA together with various classification models

```python
#here we transform X_train and X_val 
#using the PCA with 150 principal components
Z_train = pca.transform(X_train)
Z_val = pca.transform(X_val)
#yZ_train = pca.transform(y_train)
#yZ_val = pca.transform(y_val)
print(np.shape(Z_train))
print(np.shape(Z_val))
```

## Logistic Regression (PCA)

```python
%%time
pca_lr_mod = LogisticRegression(solver='lbfgs', multi_class='ovr', max_iter=6000)
pca_lr_mod.fit(Z_train, y_train)
print(pca_lr_mod.score(Z_train, y_train))
print(pca_lr_mod.score(Z_val, y_val))
```

```python
#here we use our ORIGINAL Logistic Regression model "lr_mod" trained on full 2914 features
#to predict the name (in RED) of the individual pictured with full 2914 original features
#run this cell a couple of times.  Since the VALIDATION SCORE is about .84 (see the beginning of the lecture) 
#we expect that about the SIXTH of the individuals to be recognized INCORRECTLY
i = np.random.choice(range(X_val.shape[0]))
pred = lr_mod.predict([Xs_val[i,:]])
plt.imshow(X_val[i].reshape(62,47), cmap='bone')
plt.axis('off')
plt.text(50, 5, s = names[y_val[i]], fontsize=16, color='b')
plt.text(50, 15, s = names[pred[0]], fontsize=16, color='r')
plt.show()
```

```python
#and here we use our PCA Logistic Regression model "pca_lr_mod" trained on only 150 (tranformed) features
#to recognize the individuals...   For our eyes benefit for each individual we display both 
#the original image (with 2914 features) and the reconstructed one (with only 150 features).
#Since the VALIDATED SCORE is now .80 we expect that about the FIFTH of the individuals are
#recognized INCORRECTLY.   
#Run this cell several times to see for yourself...

i = np.random.choice(range(X_val.shape[0]))
pred = pca_lr_mod.predict([Z_val[i,:]])

plt.rcParams["figure.figsize"] = [8,8]

plt.subplot(1,2,1)
plt.imshow(X_val[i].reshape(62,47), cmap='bone')
plt.axis('off')

plt.subplot(1,2,2)
pca_rep = np.sum(Z_val[i,:].reshape(150,1) * pc[:,:], axis=0).reshape(62,47)
plt.imshow(pca_rep, cmap='bone')
plt.axis('off')

plt.text(50, 5, s = names[y_val[i]], fontsize=16, color='b')
plt.text(50, 15, s = names[pred[0]], fontsize=16, color='r')

plt.show()
```

```python
pred_val = pca_lr_mod.predict(Z_val)
print( classification_report(y_val, pred_val, target_names=names))
print(confusion_matrix(y_val, pred_val))
```

## KNN (PCA)

```python
%%time
pca_KNN_mod = KNeighborsClassifier(n_neighbors=12)
pca_KNN_mod.fit(Z_train, y_train)
print(pca_KNN_mod.score(Z_train, y_train))
print(pca_KNN_mod.score(Z_val, y_val))
```

```python
pred_val = pca_KNN_mod.predict(Z_val)
print( classification_report(y_val, pred_val, target_names=names))
print(confusion_matrix(y_val, pred_val))
```

## SVM (PCA)

```python
%%time
pca_svm_mod = SVC(kernel='rbf', gamma=0.001, C=10)
pca_svm_mod.fit(Z_train, y_train)
print(pca_svm_mod.score(Z_train, y_train))
print(pca_svm_mod.score(Z_val, y_val))
```

```python
pred_val = pca_svm_mod.predict(Z_val)
print( classification_report(y_val, pred_val, target_names=names))
print(confusion_matrix(y_val, pred_val))
```

```python
#and here we use our PCA Support Vector model "pca_svm_mod" trained on only 150 (tranformed) features
#to recognize the individuals...   For our eyes benefit for each individual we display both 
#the original image (with 2914 features) and the reconstructed one (with only 150 features).
#Since the VALIDATED SCORE is now .85 we expect that only about the SIXTH of the individuals are
#recognized INCORRECTLY.   
#Run this cell several times to see for yourself...

i = np.random.choice(range(X_val.shape[0]))
pred = pca_svm_mod.predict([Z_val[i,:]])

plt.rcParams["figure.figsize"] = [8,8]

plt.subplot(1,2,1)
plt.imshow(X_val[i].reshape(62,47), cmap='bone')
plt.axis('off')

plt.subplot(1,2,2)
pca_rep = np.sum(Z_val[i,:].reshape(150,1) * pc[:,:], axis=0).reshape(62,47)
plt.imshow(pca_rep, cmap='bone')
plt.axis('off')

plt.text(50, 5, s = names[y_val[i]], fontsize=16, color='b')
plt.text(50, 15, s = names[pred[0]], fontsize=16, color='r')

plt.show()
```

In conclusion we remark that the PCA algorithm allows for a SIGNIFICANT reduction in training times with only a small deterioration of the accuracy of predictions.
